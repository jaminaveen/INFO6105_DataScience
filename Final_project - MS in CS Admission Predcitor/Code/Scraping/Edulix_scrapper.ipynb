```python
def get_student_profile_urls():
    options = Options()
    ua = UserAgent()
    userAgent = ua.random
    options.add_argument(f'user-agent={userAgent}')
    driver = webdriver.Chrome(chrome_options=options, executable_path = 'chromedriver.exe')
    driver.get('https://www.edulix.com/')
    time.sleep(50)
    driver.get('https://www.edulix.com/' + 'student-profiles')
    time.sleep(50)
    #page = requests.get("https://yocket.in/applications-admits-rejects/26228-null/2?page=" + str(i))
    profile_urls = []
    for i in range(1,19):
        html = driver.page_source
        soup = BeautifulSoup(html)
        url_containers = soup.select('#jqGrid')
        for i in url_containers[0].find_all("a", href=True):
            profile_urls.append("https://www.edulix.com"+i['href'])
        find_profiles = driver.find_element_by_xpath(r'//*[@id="next_jqGridPager"]/span')
        driver.execute_script('arguments[0].click();',find_profiles)
    return profile_urls
```
