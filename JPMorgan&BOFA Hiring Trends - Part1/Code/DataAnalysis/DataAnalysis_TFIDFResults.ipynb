```python
import pandas as pd
```

```python
combinedTFIDFresults = pd.read_csv("combinedtfidfresult.csv",low_memory=False)
```

```python
combinedTFIDFresults = combinedTFIDFresults.drop(columns="Unnamed: 0")
```

```python
combinedTFIDFresults
```

```python
sumTFIDFresults = combinedTFIDFresults.loc[:,"financial":"intelligence"].sum(axis=0)
```

```python
sumTFIDFresults
```

```python
sumTFIDFresults.to_csv("sumTFIDFresults.csv")
```

```python
keywordcountineachcompany = combinedTFIDFresults.loc[:,"financial":"intelligence"].groupby(combinedTFIDFresults["Institution"])
```

```python
keywordcountacrossalljobdescbycompany= keywordcountineachcompany.sum(axis=0)
```

```python
keywordcountacrossalljobdescbycompany
```
