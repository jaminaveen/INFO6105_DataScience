```python
import pandas as pd
```

```python
import h2o
```

```python
# Initialize local server
from h2o.estimators import H2ORandomForestEstimator

h2o.init()
```

```python
# Load data from CSV
data = h2o.import_file('../../Data/finaldataset.csv')
```

```python
data
```

```python
print(type(data))
df = pd.read_csv('../../Data/finaldataset.csv')
numerical_data = df.drop(columns=['Unnamed: 0','int_rate']).select_dtypes(include = ['int64','float'])
categorical_data = df.select_dtypes(include = ['object'])
categorical_features = categorical_data.columns.values
numerical_features = numerical_data.columns.values
categorical_features=categorical_features.tolist()
```

```python
numerical_features
```

```python
# Input parameters that are going to train
training_columns = numerical_features.tolist()
# Output parameter train against input parameters
response_column = 'int_rate'
```

```python
# Split data into train and testing
train, test = data.split_frame(ratios=[0.7])
```

```python
# Define model
'''
Model Documentation
http://h2o-release.s3.amazonaws.com/h2o/master/3420/docs-website/h2o-py/docs/modeling.html#h2orandomforestestimator
'''
model = H2ORandomForestEstimator(ntrees=50, max_depth=20, nfolds=10)
```

```python
# Train model
model.train(x=training_columns, y=response_column, training_frame=train)
```

```python
# Model performance
performance = model.model_performance(test_data=test)

print(performance)
```

```python
y_pred = model.predict(test)
```

```python
model.mae
```

#### Interpretability <br>
  - Interpretability is the degree to which a human can consistently predict the model’s result. The higher the interpretability of a machine learning model, the easier it is for someone to comprehend why certain decisions or predictions have been made. A model is better interpretable than another model if its decisions are easier for a human to comprehend than decisions from the other model
  <br>
 <b> DECISION TREE </b>
 <br>
 The interpretation is simple: Starting from the root node, you go to the next nodes and the edges tell you which subsets you are looking at. Once you reach the leaf node, the node tells you the predicted outcome. All the edges are connected by ‘AND’<br>
 - Feature importance: The overall importance of a feature in a decision tree can be computed.<br>
 - Tree decomposition:  Individual predictions of a decision tree can be explained by decomposing the decision path into one component per feature. We can track a decision through the tree and explain a prediction by the contributions added at each decision node.
