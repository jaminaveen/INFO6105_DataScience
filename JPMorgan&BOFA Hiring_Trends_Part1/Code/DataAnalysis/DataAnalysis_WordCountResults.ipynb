```python
import pandas as pd
```

```python
combinedWordCountresults = pd.read_csv("combinedWordCountresults.csv")
```

```python
combinedWordCountresults = combinedWordCountresults.drop(columns="Unnamed: 0")
```

```python
combinedWordCountresults
```

```python
sumWordCountresults = combinedWordCountresults.loc[:,"fintech":"ecosystems"].sum(axis=0)
```

```python
sumWordCountresults.to_csv("sumWordCountresults.csv")
```

```python
keywordcountineachcompany = combinedWordCountresults.loc[:,"fintech":"ecosystems"].groupby(combinedWordCountresults["Institution"])
```

```python
keywordcountacrossalljobdescbycompany= keywordcountineachcompany.sum(axis=0)
```

```python
keywordcountacrossalljobdescbycompany
```

```python
keywordcountacrossalljobdescbycompany.to_csv("WordCountResultsByCompany.csv")
```
